```python
# ==============================
# Problem Statement 1: User Login Check
# ==============================

# Given credentials
username = "admin"
password = "1234"

# User input (you can change these values to test)
input_username = "admin"
input_password = "1234"

# Checking login condition
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")


# ==============================
# Problem Statement 2: Pass / Fail Analyzer
# ==============================

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("\nPass / Fail Analysis")
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)


# ==============================
# Problem Statement 3: Simple Data Cleaner
# ==============================

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("\nCleaned Names List:")
print(cleaned_names)


# ==============================
# Problem Statement 4: Message Length Analyzer
# ==============================

messages = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Length Analysis")

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("Flag: Message longer than 10 characters")


# ==============================
# Problem Statement 5: Error Message Detector
# ==============================

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("\nTotal ERROR entries:", error_count)
```
