##Problem 1 — Smart Parking Lot Management System

```python
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: FULL - Alert!")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)

# Sample
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

##Problem 2 — Online Food Delivery Time Estimator
Logic we assume (important in interviews):

Base time = 5 min/km

• High Traffic → +20 min
• Rainy Weather → +15 min

```python
def delivery_time(distance, traffic, weather):

    eta = distance * 5  # base time

    # traffic delay
    if traffic.lower() == "high":
        eta += 20
    elif traffic.lower() == "medium":
        eta += 10

    # weather delay
    if weather.lower() == "rainy":
        eta += 15
    elif weather.lower() == "storm":
        eta += 25

    print("Estimated Delivery Time:", eta, "minutes")

# Sample
delivery_time(8, "High", "Rainy")
```

##Problem 3 — Movie Theatre Seat Occupancy Analyzer

```python
def theatre_occupancy(total_seats, booked_seats):

    occupied = len(booked_seats)
    occupancy = (occupied / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Open additional show")
    else:
        print("Show Status: Seats Available")

# Sample (150 seats booked)
booked = [1]*150
theatre_occupancy(200, booked)
```

##Problem 4 — Cloud Server Load Classification System

```python
def server_load(cpu_readings):

    total = 0
    for reading in cpu_readings:
        total += reading

    avg = total / len(cpu_readings)
    print("Average CPU Load:", round(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif 50 <= avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

# Sample
cpu = [45, 60, 70, 85, 90]
server_load(cpu)
```

##Problem 5 — Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resources):

    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

# Sample
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

##Problem 6 — Online Event Registration Capacity Controller

```python
def event_registration(capacity, registrations):

    if registrations <= capacity:
        print("Confirmed Registrations:", registrations)
        print("Waitlisted Users: 0")
        print("Registration Status: Open")
    else:
        waitlist = registrations - capacity
        print("Confirmed Registrations:", capacity)
        print("Waitlisted Users:", waitlist)
        print("Registration Status: Closed")

# Sample
event_registration(100, 105)
```
