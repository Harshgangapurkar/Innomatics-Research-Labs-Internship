
##Problem 1 — Employee Performance Bonus Eligibility

We are given employee names and performance scores in a dictionary.
The task is to:
1. Find the highest score
2. Handle ties
3. Display all employees who achieved the highest score


```python
# Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find highest score
highest_score = max(employees.values())

# Step 2: Find employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

##Problem 2 — Search Query Keyword Analysis

We process a search query string:
• Convert to lowercase
• Remove punctuation
• Count word frequency
• Display words occurring more than once


```python
import string
from collections import Counter

query = "Buy mobile phone buy phone online"

# Step 1: lowercase
query = query.lower()

# Step 2: remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: split into words
words = query.split()

# Step 4: count frequency
word_freq = Counter(words)

# Step 5: filter repeated words
repeated_keywords = {word: count for word, count in word_freq.items() if count > 1}

print(repeated_keywords)
```

##Problem 3 — Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # even numbers are valid
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

##Problem 4 — Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

from collections import Counter

# Extract domains
domains = [email.split("@")[1] for email in emails]

# Count domain usage
domain_count = Counter(domains)

total = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

##Problem 5 — Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: average sales
avg_sales = sum(sales) / len(sales)

# Step 2: detect spikes (>30% above average)
threshold = avg_sales * 1.30

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

##Problem 6 — Duplicate User ID Detection

```python
from collections import Counter

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences
id_count = Counter(user_ids)

# Filter duplicates
duplicates = {uid: count for uid, count in id_count.items() if count > 1}

# Display
for uid, count in duplicates.items():
    print(f"{uid} → {count} times")
```
