## What is OOP?

Object-Oriented Programming (OOP) is a programming paradigm that organizes code into reusable units called classes and objects.

Key Concepts Used:
• Class → Blueprint of an object
• Object → Instance of class
• Constructor (__init__) → Initializes object data
• Methods → Functions inside a class

##Problem 1 — Smart Light Controller

```python
class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"
        print(self.name, "is ON")

    def turn_off(self):
        self.status = "OFF"
        print(self.name, "is OFF")


# Object creation
light = SmartLight("Bedroom Light")
light.turn_on()
```

##Problem 2 — Employee ID Card System

```python
class Employee:

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


emp = Employee("Rahul", "EMP102", "IT")
emp.display_id()
```

##Problem 3 — Mobile Contact Record

```python
class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def save_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c1 = Contact("Anita", "9876543210")
c1.save_contact()
```

##Problem 4 — Product Price Tag Generator

```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


p = Product("Headphones", 2499)
p.price_tag()
```

##Problem 5 — Movie Rating Display System

```python
class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


m = Movie("Inception", 4.8)
m.display()
```

##Problem 6 — Delivery Address Manager

```python
class Delivery:

    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


d = Delivery("Suman", "Hyderabad")
d.delivery_details()
```
